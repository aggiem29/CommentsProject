```python
#AP trampoline building
#preparedness, collaboration, homework grades, final grade, communication
comments = []
with open("comments.csv") as c:
    header = next(c)
    for line in c:
      line = line.strip().split(',')
      comments.append(line)

      
def final_comment(name):
  info = 0
  comment = name.split()[1]
  for i in comments:
    if i[0].split()[0] == comment:
      info = i
      break
  (student, prep_string, hw, f_grade, sem_grade, asks_help, peer_collab, adjs) = (info[1], info[2], float(info[3]), float(info[4]), float(info[5]), info[6], info[7], info[8:])

  
  class_sum = "This semester in AP Trampoline Construction, we covered the history of trampolines, how trampolines have impacted the world, different trampoline building methods, ideal trampoline materials, and how to create creative trampolines. Students completed homework research assignments, hands-on in-class activities, and a final project including building a trampoline of their design.\n"
  
  
  def prep():
    sent_one = (f"{student} was {prep_string} prepared for class. You were always ready with the required materials and arrived on time.\n")
    if prep_string == "always":
      return sent_one
    if prep_string == "often":
      sent_one = sent_one.replace("always", "normally")
    if prep_string == "sometimes":
          sent_one = sent_one.replace("always", "sometimes")
    if prep_string == "rarely":
      sent_one = sent_one.replace("always", "occassionally")
    if prep_string == "never":
      sent_one = sent_one.replace(" always", "n't")
      sent_one = sent_one.replace("arrived", "didn't arrive")
    return sent_one
  
  def homework():
    if hw >= 90:
      return f"You have done a fantastic job completing the homework. Your assignments are always done creatively and artfully. Your average homework score was {hw} percent.\n"
    if hw >= 80:
      return f"You have done a good job completing the homework. Your assignments are normally done creatively (a trait highly valued in this class). Your average homework score was {hw} percent.\n"
    if hw >= 70:
      return f"You have completed most of the homework assignments. The assignments you complete are often done well, but occassionally you submit an incomplete assignment. Your average homework score was {hw} percent.\n"
    if hw >= 60:
      return f"You have completed some of the homework assignments. The assignments you complete are often done well, but there are a number of assignments you haven't submitted. Your average homework score was {hw} percent.\n"
    else:
      return f"You have not completed most of the homework assignments. You constantly turn in incomplete assignments and often turn those in late. I'm happy to help if you want some study strategies. Your average homework score was {hw} percent.\n"
      
  
  def final():
    if f_grade >= 90:
      return f"You created an amazing trampoline for your final exam! Good job thinking about the details in your trampoline. Your exam grade paired with your average homework grade has earned you a great grade in this class. Your final grade is {f_grade} percent.\n"
    if f_grade >= 80:
      return f"You created a good trampoline for your final exam! Nice job working through setbacks while constructing that trampoline. Your exam grade paired with your average homework grade has earned you a good grade in this class. Your final grade is {f_grade} percent.\n"
    if f_grade >= 70:
      return f"You created a working trampoline for your final exam. Nice job using the skills we went over in class. Your exam grade paired with your average homework grade has earned you a final grade of {f_grade} percent.\n"
    if f_grade >= 60:
      return f"You almost completed building a trampoline for your final exam. I think you could have gotten there if there was more time. Your exam grade paired with your average homework grade has earned you a final grade of {f_grade} percent.\n"
    else:
      return f"You didn't build a working trampoline for your final exam, but I appreciated your effot. Your exam grade paired with your average homework grade has earned you a final grade of {f_grade} percent.\n"
  
  def semester():
    if sem_grade >= 90:
      final_grade = "A"
    elif sem_grade >= 80:
      final_grade = "B"
    elif sem_grade >= 70:
      final_grade = "C"
    elif sem_grade >= 60:
      final_grade = "D"
    elif sem_grade < 60:
      final_grade = "F"
    return f"Your final grade for the semester is {final_grade}."
  
  def help():
    if asks_help == "yes":
      if sem_grade > 83:
        return "You have done a great job communicating with me this semester, and that has greatly contributed to their success in this class.\n"
      else:
        return "You have been diligent in communication, and while this semester presented you with some challenges, you should be proud of how you've handled them.\n"
    elif asks_help == "no":
      if sem_grade > 83:
        return "You have done a great job this semester, however, you could work on communicating with your teacher further. I am happy to offer support, but it's important that I know what and how you're doing in the class.\n"
      else:
        return "You have struggled a little to communicate with me this semester. I am happy to offer support for you in this class and help you further understand the material, but I can't do anything for you unless you talk to me, whether after class, during office hours, or free block.\n"
        
  
  def collab():
    sent_one = (f"You were {peer_collab} an active participant in group discussions. You were always contributing to peer work and worked well with your classmates.\n")
    if peer_collab == "always":
      return sent_one
    if peer_collab == "often":
      sent_one = sent_one.replace("always", "normally")
    if peer_collab == "sometimes":
          sent_one = sent_one.replace("always", "sometimes")
    if peer_collab == "rarely":
      sent_one = sent_one.replace("always", "occassionally")
    if peer_collab == "never":
      sent_one = sent_one.replace(" always", "n't")
      sent_one = sent_one.replace("well", "poorly")
    return sent_one
    
  def final_project():
    words = ""
    for adj in adjs:
      words += adj + ', '
    feedback = f"Your final project can best be described using these words: {words[:-2]}.\n"
    return feedback

 

  with open(f"{comment.lower()}.txt", "w") as fin:
    header = f"{name}\n\n"
    fin.write(header)
    fin.write(class_sum)
    fin.write(prep())
    fin.write(homework())
    fin.write(collab())
    fin.write(help())
    fin.write(final_project())
    fin.write(final())
    fin.write(semester())

  return f"{comment.lower()}.txt"
      
with open(final_comment("Rahul Vanni")) as c:
  for line in c:
    print(line)
```
